```python
import imageio

import matplotlib.pyplot as plt
```

```python

# image 1
im_1_image=imageio.imread("Mini_BAGLS_dataset/0.png")

# im_1_meta=imageio.imread("Mini_BAGLS_dataset/0.meta")
# plt.imshow(im_1_meta)
import json

with open("Mini_BAGLS_dataset/0.meta", "r") as file:
    data = json.load(file)
    print(data)

im_1_seg=imageio.imread("Mini_BAGLS_dataset/0_seg.png")

# showing images
fig, axes = plt.subplots(1, 2, figsize=(6, 4))
axes[0].imshow(im_1_image)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(im_1_seg)
axes[1].set_title("Segmented Image")
axes[1].axis('off')

plt.tight_layout()
plt.show()


```
