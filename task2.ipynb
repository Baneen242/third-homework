```python
import imageio.v3 as imageio
import matplotlib.pyplot as plt
import json

# List of image indices
image_indices = [0, 1, 2, 3]

# Create a figure with subplots for all images
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

# Iterate over image indices to load and plot images and their segmentation masks
for i, idx in enumerate(image_indices):
    # Load image, segmentation mask, and metadata
    im_image = imageio.imread(f"Mini_BAGLS_dataset/{idx}.png")
    im_seg = imageio.imread(f"Mini_BAGLS_dataset/{idx}_seg.png")
    
    with open(f"Mini_BAGLS_dataset/{idx}.meta", "r") as file:
        data = json.load(file)
        disorder_status = data.get("Subject disorder status", "Unknown")
    
    # Display original image
    axes[0, i].imshow(im_image)
    axes[0, i].set_title(f"Original Image\nStatus: {disorder_status}")
    axes[0, i].axis('off')

    # Display segmentation mask overlayed on the original image
    axes[1, i].imshow(im_image)
    axes[1, i].imshow(im_seg, cmap='jet', alpha=0.5)  # Overlay the mask with transparency
    axes[1, i].set_title(f"Overlayed Image\nStatus: {disorder_status}")
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```
